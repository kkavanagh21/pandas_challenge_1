## Part 1: Explore the Data

Import the data and use Pandas to learn more about the dataset.

```python

import pandas as pd

# Load the dataset
file_path = 'client_dataset.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()


```

```python
# View the column names in the data
column_names = data.columns
print("Column Names:", column_names)

# Use the describe function to gather basic statistics
basic_stats = data.describe()
print("Basic Statistics:\n", basic_stats)
```

```python
# Use this space to do any additional research
# and familiarize yourself with the data.
```

```python
# What three item categories had the most entries?
top_categories = data['category'].value_counts().head(3)
print("Top 3 Categories with Most Entries:\n", top_categories)
```

```python
# For the category with the most entries, which subcategory had the most entries?
top_category = top_categories.index[0]

# Filter data for the top category and find the top subcategory
top_subcategory = data[data['category'] == top_category]['subcategory'].value_counts().idxmax()
print(f"Subcategory with most entries in {top_category} category: {top_subcategory}")
```

```python
# Which five clients had the most entries in the data?
top_clients = data['client_id'].value_counts().head(5)
print("Top 5 Clients with Most Entries:\n", top_clients)
```

```python
# Store the client ids of those top 5 clients in a list.
top_clients_list = top_clients.index.tolist()
print("Top 5 Client IDs:", top_clients_list)
```

```python
# How many total units (the qty column) did the client with the most entries order order?
top_client_id = top_clients.index[0]
total_units_top_client = data[data['client_id'] == top_client_id]['qty'].sum()
print(f"Total units ordered by the top client (ID: {top_client_id}): {total_units_top_client}")
```

## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
data['subtotal'] = data['unit_price'] * data['qty']
```

```python
data.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

data['shipping_price'] =(data['unit_weight'] * data['qty']).apply(lambda x: 7 * x if x > 50 else 10 * x)
data.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
data['total_price'] = (data['subtotal'] + data['shipping_price']) * 1.0925
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
data['line_cost'] = (data['unit_cost'] * data['qty']) + data['shipping_price']

data.head()
```

```python
# Create a column for the profit of each line using line cost and line price
data['profit'] = data['total_price'] - data['line_cost']

data.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
top_clients_spending = data[data['client_id'].isin(top_clients_list)].groupby('client_id')['total_price'].sum()
print("Total spending by top 5 clients:\n", top_clients_spending)

```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
summary_df = data[data['client_id'].isin(top_clients_list)].groupby('client_id').agg({
    'qty': 'sum',
    'shipping_price': 'sum',
    'total_price': 'sum',
    'profit': 'sum'
}).rename(columns={
    'qty': 'Total Units Purchased',
    'shipping_price': 'Total Shipping Price',
    'total_price': 'Total Revenue',
    'profit': 'Total Profit'
}).sort_values(by='Total Profit', ascending=False)

print("Summary DataFrame:\n", summary_df)
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 

# Define a function that converts a dollar amount to millions.
def to_millions(x):
    return x / 1_000_000

formatted_summary_df = summary_df.applymap(to_millions) 
#Apply the currency_format_millions function to only the money columns. 
formatted_summary_df.columns = [col + ' (in millions)' for col in formatted_summary_df.columns]

# Rename the columns to reflect the change in the money format. 
formatted_summary_df.rename(columns={
    'Total Units Purchased (in millions)': 'Total Units (in millions)',
    'Total Shipping Price (in millions)': 'Shipping Cost (in millions)',
    'Total Revenue (in millions)': 'Revenue (in millions)',
    'Total Profit (in millions)': 'Profit (in millions)'
}, inplace=True)

formatted_summary_df.reset_index(inplace=True)
formatted_summary_df.rename(columns={'client_id': 'Client ID'}, inplace=True)

print("Formatted Summary DataFrame:\n", formatted_summary_df)


```
